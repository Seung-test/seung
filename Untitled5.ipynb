```python
import pandas as pd
movies = pd.read_csv("/content/movies.csv")
print(movies)
```

```python
genre_movies = movies[["title", "genre"]]
print(genre_movies)
```

```python
def recommend_movie(movie_title):

  genre = movies[movies["title"] == movie_title]["genre"].values[0]

  recommend = movies[movies["genre"] == genre]

  return recommend

print(recommend_movie("Inception"))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

movie_vector = np.array([
    [0,1,0], # Inception
    [0,1,0], # Interstellar
    [1,0,0], # Avengers
    [0,0,1], # Titanic
])


similarity = cosine_similarity(movie_vector)
print(similarity)
```
